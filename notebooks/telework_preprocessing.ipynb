### Clean `us_df`

```python
us_df.rename(columns={'Area': 'Country',
                      'Number': 'Number US',
                      'MOE': 'MOE US',
                      'Percent': 'Percent US',
                      'Percent MOE': 'Percent MOE US'}, 
             inplace=True)

us_df.sample(5, random_state=24)
```

### Clean `states_df`

```python
states_df.rename(columns={'Area': 'State',
                          'Number': 'Number State',
                          'MOE': 'MOE State',
                          'Percent': 'Percent State',
                          'Percent MOE': 'Percent MOE State'},
                 inplace=True)

states_df.sample(5, random_state=24)
```

### Clean `dc_df`

```python
dc_df.rename(columns={'Area': 'State',
                      'Number': 'Number City',
                      'MOE': 'MOE City',
                      'Percent': 'Percent City',
                      'Percent MOE': 'Percent MOE City'},
             inplace=True)

dc_df['City'] = np.where(dc_df['State'].str.contains('District of Columbia'), 'Washington', np.nan)

dc_df = dc_df[['City', 'State', 'State Abbr', 'Number City', 
               'MOE City', 'Percent City', 'Percent MOE City', 
               'Date', 'Month', 'Year', 'Week Num']]
dc_df.sample(5, random_state=24)
```

### Clean `city_df`

```python
city_df.rename(columns={'Area': 'Metro Area',
                        'Number': 'Number City',
                        'MOE': 'MOE City',
                        'Percent': 'Percent City',
                        'Percent MOE': 'Percent MOE City'},
                 inplace=True)

city_df.sample(5, random_state=24)
```

```python
city_df['Metro Area'] = city_df['Metro Area'].str.split('Metro Area', expand=True)[0]
city_df[['Metro Area', 'State Abbr']] = city_df['Metro Area'].str.split(',', expand=True)
```

```python
city_df.sample(5, random_state=24)
```

```python
city_df[['State Abbr', 'State Abbr2', 'State Abbr3', 'State Abbr4']] = city_df['State Abbr'].str.split('-', expand=True)
```

```python
city_df = pd.melt(city_df, 
               id_vars=['Metro Area', 'Date', 'Number City', 'MOE City', 
                        'Percent City', 'Percent MOE City', 'Week Num',
                        'Month', 'Year'], 
               value_vars=['State Abbr', 'State Abbr2', 'State Abbr3', 'State Abbr4'],
               value_name='State Abbr'
              ).dropna()

city_df.drop(['variable'], axis=1, inplace=True) 
```

```python
city_df.sample(5, random_state=24)
```

```python
city_df[['City 1', 'City 2', 'City 3']] = city_df['Metro Area'].str.split('-', expand=True)
```
